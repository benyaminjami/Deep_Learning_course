# Network Structure

```python
import tensorflow as tf
tf.reset_default_graph()
BatchSize = 64

def D_logit(inp): # construct discriminator structure
    last = tf.reshape(inp,(BatchSize,784))
    last = tf.layers.dropout(last,training=True,rate=0.5)
    for inx, size in enumerate([256,128,64]):
        last = tf.layers.dense(last,size,activation=tf.nn.leaky_relu,name="L%d"%inx)
    last = tf.layers.dropout(last,training=True,rate=0.5)
    logit = tf.layers.dense(last,1,activation=None,name="Final")
    return logit
    
def G(z): # construct generator structure
    last = z
    for inx, size in enumerate([128,256,512]):
        last = tf.layers.dense(last,size,activation=tf.nn.leaky_relu,name="L%d"%inx)
    last = tf.layers.dense(last,784,activation=tf.nn.sigmoid,name="Final")
    img = tf.reshape(last,(BatchSize,28,28))
    return img

Z = tf.placeholder(tf.float32, [BatchSize, 128], name="latent_value")
with tf.variable_scope("Generator"):
    fake_img = G(Z)
    
real_img = tf.placeholder(tf.float32, [BatchSize, 28, 28], name="real_image")
with tf.variable_scope("Discriminator"):
    D_logit_real = D_logit(real_img)
with tf.variable_scope("Discriminator", reuse=True):
    D_logit_fake = D_logit(fake_img)
```

# Loss functions

```python
D_loss_real = tf.reduce_mean(
    tf.nn.sigmoid_cross_entropy_with_logits(logits=D_logit_real, labels=tf.ones_like(D_logit_real))
)
D_loss_fake = tf.reduce_mean(
    tf.nn.sigmoid_cross_entropy_with_logits(logits=D_logit_fake, labels=tf.zeros_like(D_logit_fake))
)
D_loss = D_loss_real+D_loss_fake

G_loss = tf.reduce_mean(
    tf.nn.sigmoid_cross_entropy_with_logits(logits=D_logit_fake, labels=tf.ones_like(D_logit_fake))
)
```

# Optimizer

```python
D_optim_vars = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope="Discriminator")
print("D tensor number:", len(D_optim_vars))
G_optim_vars = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, scope="Generator")
print("G tensor number:", len(G_optim_vars))

D_optim = tf.train.AdamOptimizer(learning_rate=2e-4)
D_train_op = D_optim.minimize(D_loss, var_list=D_optim_vars)

G_optim = tf.train.AdamOptimizer(learning_rate=2e-4)
G_train_op = G_optim.minimize(G_loss, var_list=G_optim_vars)
```

# Dataset

# Training

```python
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)

sess.run(tf.global_variables_initializer())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 3) # set default size of plots

def Z_Generator():
    return np.random.normal(size=(BatchSize,128))

MasterZ = Z_Generator()
for step in range(10**5):
    # train D
    for _ in range(5):
        data, _ = mnist.train.next_batch(BatchSize)
        data = data.reshape(BatchSize,28,28)
        sess.run(D_train_op,feed_dict={real_img:data, Z:Z_Generator()})
        
    # train G
    sess.run(G_train_op, feed_dict={Z:Z_Generator()})
    
    if step%5000 ==0 :
        print("="*30 + str(step) + "="*30)
        G_loss_value, D_loss_value = sess.run([G_loss,D_loss],feed_dict={real_img:data, Z:Z_Generator()})
        print("D loss = %f\tG loss=%f"%(D_loss_value,G_loss_value))
        
        images = sess.run(fake_img,feed_dict={Z:MasterZ})
        # plot the images
        for i in range(3):
            for j in range(10):
                plt.subplot(3, 10, i * 10 + j + 1)
                plt.imshow(images[i * 10 + j], cmap="gray", vmin=0, vmax=1)
        plt.show()
```

# Interpolation in latent space

```python
import matplotlib.pyplot as plt
%matplotlib inline
interpolate_num = 20
img_num = 15
plt.rcParams['figure.figsize'] = (img_num, interpolate_num) # set default size of plots
SourceZ = Z_Generator()
TargetZ = Z_Generator()

diff = (TargetZ-SourceZ)/float(interpolate_num-1)

tmp = SourceZ
for i in range(interpolate_num):
    tmp += diff
    images = sess.run(fake_img,feed_dict={Z:tmp})
    for j in range(img_num):
        plt.subplot(interpolate_num, img_num, i*img_num + j +1)
        plt.imshow(images[j], cmap="gray", vmin=0, vmax=1)
plt.show()
```
